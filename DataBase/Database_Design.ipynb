This project has two phases:

1. Data Entry Phase
2. Roll Out Phase

The databases and user tool will be different in the different phases. In many ways the purpose of phase I is to prepare the way for phase II, and the structural differences between the two resulting tools may be small. However, whilst I am able to keep in mind the end goal, I will compartmentalise the two projects somewhat separately. It is not practical in terms of programming time nor data entry time to enter data needed for phase II but not phase I during the Data Entry Phase. Additionally a duality of goals will only lead to confusion in the strucutre. 

To me this seems a separation of concerns issue: http://goo.gl/1DVRbu

Therefore in this document I will set out what is needed to achieve the Data Entry Phase. However, I am keenly aware that part of my mandate is to create something that the Association will want to use itself (i.e. Phase II). Therefore at all times I will try to note down the extra steps we would want to take for the Roll Out Phase.

##Entity Modelling<br/>
##------------------------
###MEMBER
The first entity that we will consider for the database is the MEMBER. A MEMBER is a corporate or other entity that is a member of the Myanmar Garment Manufacturers Association. At present I am working on the basis of the information provided to me in the form of the 'Member List'. 

The MEMBER set has the following attributes:

| | | | | 
|---|---|---|---|
|**Attribute** | **Domain** | **Constraints**| **Notes** |
| mem_id | string | PRIMARY KEY | Automatically generated
| mem_name | string | NOT NULL |  n/a  |
| mem_address | string | NOT NULL | n/a  |

The above are the attributes that will be present in the Data Collection Tool.

There are some notable absences from the above. For example I do not include:

+ contact person
+ telephone number
+ fax number

Those things will however be needed for the Roll Out Phase. The reason for not including them in Phase I is that they have no bearing on the data we are collecting, but more importantly that these things are liable to change over time, and possible even within the time period of data that we will be collecting. As we have no need for the 'most up to date' information, there is no point putting in funcationality that allows the user to modify records, and if there is no need for that, then there is no need to have the information in the first place. 

Of course the Phase II tool will need this info, but I would suggest that proper palce for it would be as an attrbute of the ORDER entity - i.e. the number and name is provided with each order, therefore making it uncessary to update the central database every time the name/number changes. 

There is one other attribute that might be required for Phase I and Phase II and that is the 'Investment Type' There are issues here though: see 'Points for Resolution' (below).

###INPUT
The fourth entity that we will have in the database is the INPUT. An INPUT is an item of apparel related input that must be imported in order to create the ITEM. 

The interpretation of INPUT is very tricky actually, and hence I wait for further information/guideance. On the one hand we might think of the INPUT as being related to the ORDER, or we might think think of the INPUT being realted to the ITEM. It is not easy for these interpretations to coexist side by side. 

In one reality each INPUT is related only to the ORDER. This is the simplest case. However, we have also seen that on one form the INPUT is apportioned between the ITEMs. In this case INPUT is realted to the ITEM. But what we dont know is how completely the INPUTs are apporitioned to the ITEMs (for example on the second form it seems that there is no approtionment given). If the INPUT is not perfectly apportioned to the ITEM then we will generate misleading data if the link is between ITEM and INPUT rather than between ORDER and INPUT.
